```python
import pandas as pd
import numpy as np
import neattext.functions as nfx
import seaborn as sns
```

```python
df=pd.read_csv('udemy.csv')
df.head()
```

```python
df.dropna(subset=['course title'], inplace=True)
```

```python
df['clean_title'] = df['course title'].astype(str).apply(nfx.remove_stopwords)
df['clean_title']=df['clean_title'].astype(str).apply(nfx.remove_special_characters)
df['clean_title'].iloc[1:5]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
```

```python
cv=CountVectorizer()
cv_mat=cv.fit_transform(df['clean_title'])
cv_mat
```

```python
cv_mat.todense()
```

```python
df_cv=pd.DataFrame(cv_mat.todense(),columns=cv.get_feature_names_out())
df_cv
```

```python
cosine_sim_mat=cosine_similarity(cv_mat)
cosine_sim_mat
```

```python
def recommend_course(title,numrec=10):
  course_index=pd.Series(df.index,index=df['course title']).drop_duplicates()

  index = int(course_index[title])
  scores = list(enumerate(cosine_sim_mat[index]))
  sorted_score=sorted(scores,key=lambda x:x[1],reverse=True)
  selected_course_index=[i[0] for i in sorted_score[1:]]
  selected_course_score=[i[1] for i in sorted_score[1:]]
  rec_df=df.iloc[selected_course_index]
  rec_df['similarity_score']=selected_course_score
  final_recommended_courses=rec_df[['course title','similarity_score','course url','Ratings']]
  return final_recommended_courses.head(numrec)
```

```python
ans = recommend_course('Deep Learning: Convolutional Neural Networks in Python',20)

ans
```
